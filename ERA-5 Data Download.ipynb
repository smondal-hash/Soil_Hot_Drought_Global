```python
import ee
import datetime

# Initialize the Earth Engine
ee.Initialize()

# Define the dataset
collection = ee.ImageCollection("ECMWF/ERA5_LAND/HOURLY")

# Define the date range
start_year = 1979
end_year = 2023

# Define the region (Conterminous United States)
conus_bounds = ee.Geometry.Rectangle([-125, 24, -66, 50])  # Adjust these coordinates as needed

# Function to export the first hourly image for a given year
def export_first_hourly_image(year):
    start_date = datetime.datetime(year, 1, 1)
    end_date = start_date.replace(year=year + 1)
    
    # Filter collection for the specified year and select the 'total_precipitation' band
    yearly_collection = collection.filterDate(start_date.isoformat(), end_date.isoformat()).select('total_precipitation')

    # Get the first image
    first_image = yearly_collection.first()

    # Define export parameters
    export_params = {
        'description': f'ERA5_LAND_precipitation_hourly_{year}',
        'folder': 'ERA5_LAND_Precipitation_Hourly',  # Google Drive folder name
        'scale': 10000,  # Resolution, adjust as needed
        'region': conus_bounds,  # Region of interest
        'fileFormat': 'GeoTIFF',  # Export format
    }
    
    # Start the export task
    task = ee.batch.Export.image.toDrive(first_image, **export_params)
    task.start()

    print(f"Export started for the first image of {year}")

# Loop over each year and start the export
for year in range(start_year, end_year + 1):
    export_first_hourly_image(year)

print("All export tasks for the first hourly image of each year started.")

```
